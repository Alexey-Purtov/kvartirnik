```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('_data.csv')
```

```python
pd.set_option('display.max_columns', None) # чтобы увидеть все столбцы
```

```python
df
```

```python
# Фильтруем строки с адресами, содержащими слово "Москва"
moscow_count = df["Адрес"].str.contains("Москва").sum()
not_moscow_count = (~df["Адрес"].str.contains("Москва")).sum()

# Данные для графика
counts = [moscow_count, not_moscow_count]
labels = ['Квартиры в Москве', 'Квартиры не в Москве']

# Создаем график
plt.bar(labels, counts, color=['blue', 'orange'])
plt.title('Количество квартир в Москве и не в Москве')
plt.ylabel('Количество квартир')
plt.xlabel('Категории')
plt.xticks(rotation=15)
plt.tight_layout()
```

```python
# Фильтрация строк, содержащих слово "Москва"
df = df[df['Адрес'].str.contains('Москва', case=False, na=False)].reset_index(drop=True)
```

```python
df
```

```python
# Анализ по колнкам: сколько и каких вариаций, пропущенных занчений (NaN), типы данных
numb = 1
feature_names = df.columns.tolist() 
for column in feature_names: 
    print('===============', numb, 'колонка', '===============')
    print('Кол-во вариаций',df[column].nunique())
    print('==========================================')
    print(df[column].value_counts(dropna=False).sort_values())
    print()
    numb += 1
```

```python
missing_data = df.isnull().sum()  # Количество NaN (пустых) значений
filled_data = df.notnull().sum()  # Количество заполненных значений

#Создание DataFrame для удобства
data_summary = pd.DataFrame({
    'Пустые значения': missing_data,
    'Заполненные значения': filled_data
})

#Построение stacked bar chart (составная диаграмма)
plt.figure(figsize=(14, 10))
data_summary.plot(kind='bar', stacked=True, color=['red', 'green'], figsize=(12, 6))

#Настройка графика
plt.title('Соотношение пустых и заполненных значений в колонках', fontsize=16)
plt.xlabel('Колонки', fontsize=12)
plt.ylabel('Количество значений', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.legend(['Пустые значения', 'Заполненные значения'], fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Отображение графика
plt.show()
```

# Анализ целевой переменной #

```python
df['Цена'].sort_values() # есть доллары и евро
```

```python
def convert(n):
    parts = n.split()
    if '$' in parts[1]:
        return float(parts[0]) * 90  # Конвертация долларов в рубли
    elif '€' in parts[1]:
        return float(parts[0]) * 100  # Конвертация евро в рубли
    else:
        return float(parts[0])  # Если рубли, просто возвращаем значение

# Применение функции к столбцу 'Цена'
df['Цена'] = df['Цена'].map(convert)
```

```python
# Визуализация распределения цены
plt.figure(figsize=(10, 6))
sns.histplot(df['Цена'], bins=50, kde=True)
plt.title('Распределение цены аренды')
plt.xlabel('Цена (руб.)')
plt.ylabel('Частота')
```

# Анализ числовых признаков #

```python
df['Площадь, м2'] # надо оставить только общую площадь
```

```python
# Функция для извлечения общей площади
def extract_total_area(area):
    parts = area.split('/')
    return float(parts[0]) if len(parts) > 0 else 0

# Применение функции к столбцу и создание нового столбца с общей площадью
df['Площадь, м2'] = df['Площадь, м2'].apply(extract_total_area)
```

```python
# Визуализация корреляции между площадью и ценой
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Площадь, м2', y='Цена', data=df)
plt.title('Зависимость цены от общей площади')
plt.xlabel('Общая площадь (м²)')
plt.ylabel('Цена (руб.)')
```
